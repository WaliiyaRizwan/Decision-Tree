```python
import numpy as np                          #works with array data sets
import pandas as pd                         #offers data structures and operations for manipulating numerical tables and time series
from pandas import DataFrame, Series
from IPython.display import Image           #Create a PNG/JPEG/GIF image object given raw data.
from io import StringIO                     #for file creation
import pydotplus                            #provides a Python Interface to Graphviz's Dot language
from sklearn import preprocessing           #classes to change raw feature vectors into a representation that is more suitable for the downstream estimators.
from sklearn import tree                    #to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features
%matplotlib inline                          #%matplotlib inline sets the backend of matplotlib to the 'inline' backend
```

```python
def convert_yes_no(txt):
    if 'yes' in txt:
        return 1
    else:
        return 0
```

```python
def plot_decision_tree(clf,feature_name,target_name):
    dot_data = StringIO()    #creats empty file
    tree.export_graphviz(clf, out_file=dot_data,      #this function generates a GraphViz representation of the decision tree, which is then written into out_file
                         feature_names=feature_name,  
                         class_names=target_name,  
                         filled=True, rounded=True,  
                         special_characters=True)  
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
    return Image(graph.create_png())
```

```python
df = pd.read_csv('tennis.csv')
```

```python
df
```

```python
df.windy = df.windy.astype(int)
df.play = df.play.apply(convert_yes_no)
```

```python
df
```

```python
df = pd.get_dummies(df) #It converts categorical data into dummy or indicator variables. Parameters: data: whose data is to be manipulated.
df
```

```python
# attrubutes / features
#. loc[] method is a method that takes only index labels and
#returns row or dataframe if the index label exists in the caller data frame
X_train = df.loc[:, df.columns != 'play'] # all rows and columns except play column
```

```python
#target
Y_train = df.play
```

```python
#The function to measure the quality of a split.
#Supported criteria are “gini” for the Gini impurity and “entropy” for the information gain.
#clf is a classifier defined as an object for our model
clf = tree.DecisionTreeClassifier(criterion='entropy') #here entropy is our Attribute measure
```

```python
#The fit method fits the training dataset as features (data) and labels (target) into the model
clf = clf.fit(X_train,Y_train)
```

```python
#visulaizing the decision tree
plot_decision_tree(clf, X_train.columns,df.columns[1])
```
